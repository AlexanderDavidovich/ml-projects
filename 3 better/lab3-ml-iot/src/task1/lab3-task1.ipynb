## Часть 1: Примеры команд awk

### Пример 1: Вывод файла со списком студентов

```python
# Просмотр содержимого файла
!cat ../data/list_students.txt
```

### Пример 2: Вывод фамилий, имен и рейтинговых оценок

```python
# Вывод полей 1, 2 и 5 (фамилия, имя, оценка)
!awk '{print $1, $2, $5}' ../data/list_students.txt
```

### Пример 3: Вывод всего файла с помощью переменной $0

```python
# Переменная $0 представляет всю строку
!awk '{print $0}' ../data/list_students.txt
```

### Пример 4: Поиск студентов с рейтинговой оценкой 5

```python
# Поиск строк, содержащих '5'
!awk '/5/ {print $0}' ../data/list_students.txt
```

### Пример 5: Точный поиск по полю (рейтинг == 5)

```python
# Точное совпадение 5-го поля (рейтинг) со значением 5
!awk '($5 == 5)' ../data/list_students.txt
```

### Пример 6: Выборка студентов факультета ЭКТ, 1 курса

```python
# Условие с логическим И (&&)
!awk '($3 == "ЭКТ") && ($4 == 1) {print}' ../data/list_students.txt
```

## Часть 2: Примеры команд sed

### Пример 1: Просмотр файла books

```python
!cat ../data/books
```

### Пример 2: Замена слова в файле

```python
# Замена 'Яшин' на 'Кашин'
!sed 's/Яшин/Кашин/' ../data/list_students.txt
```

### Пример 3: Замена 'ЭКТ' на 'КБ'

```python
!sed 's/ЭКТ/КБ/' ../data/list_students.txt
```

### Пример 4: Удаление строк с 3 по 6

```python
!sed '3,6d' ../data/list_students.txt
```

### Пример 5: Удаление строк, содержащих 'ЭКТ'

```python
!sed '/ЭКТ/d' ../data/list_students.txt
```

### Пример 6: Вывод строк, содержащих 'book'

```python
!sed -n '/book/ p' ../data/books
```
