## Пример 1: Работа с файлом myfile

```python
# Создание файла myfile
!cat ../data/myfile
```

```python
# Вывод первого элемента каждой строки
!awk '{print $1}' ../data/myfile
```

```python
# Вывод третьего элемента каждой строки
!awk '{print $3}' ../data/myfile
```

## Пример 2: Преобразование регистра

```python
# Создание тестового файла для примера 2
test_content = """This line contains bar.
This line contains foo."""

with open('../data/test_toupper.txt', 'w') as f:
    f.write(test_content)

# Преобразование строк с 'foo' в верхний регистр
!awk '/foo/ { print toupper($0); }' ../data/test_toupper.txt
```

## Пример 3: Работа со списком студентов

```python
# Создание файла list_students.txt с расширенным списком
students_content = """Асташко Иван ПИ 1 6
Бузун Евгений КБ 1 9
Кравченя Леонид ПМ 1 4
Кошкин Владимир ИН 1 7
Липин Федор ПИ 2 4
Пенов Николай ПМ 1 5
Яшин Петр КБ 4 8"""

with open('../data/list_students_new.txt', 'w') as f:
    f.write(students_content)

# Вывод списка
!cat ../data/list_students_new.txt
```

```python
# Вывод фамилий, имен и рейтинговых оценок
!awk '{print $1,$2,$5}' ../data/list_students_new.txt
```

## Пример 4: Выборка студентов с рейтингом 8

```python
!awk '/8/ {print $0}' ../data/list_students_new.txt
```

## Пример 5: Точная выборка по рейтингу

```python
# Выбор студентов с рейтингом ровно 4
!awk '($5==4)' ../data/list_students_new.txt
```

## Пример 6: Выборка по нескольким условиям

```python
# Студенты специальности КБ, 1 курса
!awk '($3=="КБ")&&($4==1)' ../data/list_students_new.txt
```

## Пример 7: Длина строки

```python
# Вывод студентов с именем длиннее 5 символов
!awk '(length($2)>5) {print}' ../data/list_students_new.txt
```

## Пример 8: Нумерация строк

```python
# Вывод с номерами строк
!awk '{print NR, $0}' ../data/list_students_new.txt
```

## Пример 9: Преобразование в верхний регистр

```python
# Вывод имен в верхнем регистре
!awk '{print toupper ($2)}' ../data/list_students_new.txt
```

## Пример 10: Подсчет суммы

```python
# Суммарный балл оценок
!awk '{sum += $5} END {print("SUM=",sum)}' ../data/list_students_new.txt
```

## Пример 11: Поиск по регулярному выражению

```python
# Проверка наличия фамилии Кравченя
!awk '($1 ~ /Кравченя/)' ../data/list_students_new.txt
```

## Пример 12: Регистронезависимый поиск

```python
# Поиск Липин (с любым регистром первой буквы)
!awk '($1 ~ /[Лл]ипин/)' ../data/list_students_new.txt
```

```python
# Вывод всех КРОМЕ Липин
!awk '!($1 ~ /[Лл]ипин/)' ../data/list_students_new.txt
```

## Пример 13: Работа с CSV файлом

```python
# Просмотр CSV файла
!cat ../data/colours.csv
```

```python
# Выборка записей с количеством больше 5
!awk -F"," '$3>5 {print $1, $2}' ../data/colours.csv
```
