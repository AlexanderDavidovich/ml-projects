## Задача 1: Поиск строк с "Network Protocol" в RFC файле

Напишите awk-программу, которая читает данные из файла https://www.rfc-editor.org/rfc/rfc-ref.txt, выводит строки с фразой «Network Protocol» и подсчитывает их общее количество.

```python
import urllib.request
import os

# Скачивание файла (если еще не скачан)
file_path = '../data/rfc-ref.txt'
if not os.path.exists(file_path):
    url = 'https://www.rfc-editor.org/rfc/rfc-ref.txt'
    print("Скачивание файла...")
    urllib.request.urlretrieve(url, file_path)
    print("Файл скачан!")
else:
    print("Файл уже существует")
```

```python
# AWK программа для поиска "Network Protocol"
!awk '/Network Protocol/ {print NR": "$0} END {print "\nВсего найдено строк:", NR-1}' ../data/rfc-ref.txt | grep "Network Protocol"
```

```python
# Более точный подсчет с выводом результата
!awk '/Network Protocol/ {count++; print NR": "$0} END {print "\n===================="; print "Общее количество строк с 'Network Protocol':", count}' ../data/rfc-ref.txt
```

## Задача 2: Расчет площади и периметра параллелограмма

Создайте скрипт awk, состоящий из двух функций для расчета площади и периметра параллелограмма и которые вызываются из функции main.

Формулы:
- Площадь: S = a × h (где a - основание, h - высота)
- Периметр: P = 2(a + b) (где a и b - стороны)

```python
# Создание AWK скрипта для параллелограмма
awk_script = '''function area(base, height) {
    return base * height
}

function perimeter(side_a, side_b) {
    return 2 * (side_a + side_b)
}

function main(a, b, h) {
    s = area(a, h)
    p = perimeter(a, b)
    
    print "==================================="
    print "Параметры параллелограмма:"
    print "  Основание (a):", a
    print "  Боковая сторона (b):", b
    print "  Высота (h):", h
    print "==================================="
    print "Результаты расчетов:"
    print "  Площадь (S = a × h):", s
    print "  Периметр (P = 2(a + b)):", p
    print "==================================="
}

BEGIN {
    # Пример: основание = 10, боковая сторона = 6, высота = 5
    main(10, 6, 5)
    
    print "\\n"
    
    # Еще один пример
    main(15, 8, 7)
}'''

with open('../data/parallelogram.awk', 'w') as f:
    f.write(awk_script)

print("AWK скрипт создан: ../data/parallelogram.awk")
```

```python
# Запуск AWK скрипта
!awk -f ../data/parallelogram.awk
```

## Задача 3: Замена пробела на 0 перед числом

Напишите sed-команду, копирующую файл на стандартный вывод, заменяя при этом одиночный пробел, являющийся первым символом в строке, нулем (0) только в том случае, если за пробелом сразу же следует число (0–9).

```python
# Просмотр тестового файла
!cat ../data/test_spaces.txt
```

```python
# SED команда для замены начального пробела на 0 перед числом
!sed 's/^ \([0-9]\)/0\1/' ../data/test_spaces.txt
```

### Объяснение sed команды:

- `s/` - начало команды замены
- `^` - начало строки
- ` ` - одиночный пробел
- `\([0-9]\)` - группа захвата: любая цифра от 0 до 9
- `/0\1/` - замена: 0 + захваченная цифра (\1)

Результат: строки вида " 85c" и " 123" преобразуются в "085c" и "0123"

```python
# Создание еще одного тестового файла для демонстрации
test_content = """abc abc
abc abc
 85c 085c
b5b b5b
 000 0000
 99 test
normal line
 text no digit"""

with open('../data/test_spaces2.txt', 'w') as f:
    f.write(test_content)

print("Исходный файл:")
!cat ../data/test_spaces2.txt
```

```python
print("\nПосле применения sed:")
!sed 's/^ \([0-9]\)/0\1/' ../data/test_spaces2.txt
```
