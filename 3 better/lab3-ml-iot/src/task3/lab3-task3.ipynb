## Пример 1: Вывод всех строк и дублирование по шаблону

```python
# Просмотр файла books
!cat ../data/books
```

```python
# Дублирование строк, содержащих 'book'
!sed '/book/ p' ../data/books
```

## Пример 2: Выбор строк по шаблону

```python
# Вывод только строк, содержащих 'book'
!sed -n '/book/ p' ../data/books
```

## Пример 3: Вывод диапазона строк

```python
# Вывод строк со 2 по 5
!sed -n '2,5 p' ../data/books
```

## Пример 4: Использование файла с командами

```python
# Создание файла с командами sed
with open('../data/records', 'w') as f:
    f.write('2,5 p\n')

# Выполнение команд из файла
!sed -n -f ../data/records ../data/books
```

## Пример 5: Добавление текста после строки

```python
# Создание файла с командой добавления
with open('../data/appends', 'w') as f:
    f.write('3 a\\\nMy favorite book.\n')

# Добавление текста после 3-й строки
!sed -f ../data/appends ../data/books
```

## Пример 6: Вставка текста перед строкой

```python
# Создание файла с командой вставки
with open('../data/insert', 'w') as f:
    f.write('/This/ i\\\nSKARBONKA.\n')

# Вставка текста перед строками с 'This'
!sed -f ../data/insert ../data/books
```

## Пример 7: Замена слов

```python
# Замена 'book' на 'novel'
!sed -n 's/book/novel/ p' ../data/books
```

## Пример 8: Удаление строк по диапазону

```python
# Удаление строк с 3 по 6
!sed '3,6d' ../data/list_students.txt
```

## Пример 9: Удаление по шаблону

```python
# Удаление строк, содержащих 'ЭКТ'
!sed '/ЭКТ/d' ../data/list_students.txt
```

## Пример 10: Множественные замены

```python
# Глобальная замена всех вхождений
!echo 'book book book' | sed 's/book/novel/g'
```
