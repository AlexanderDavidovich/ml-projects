# Ответы на вопросы по метрикам классификации

## 1. Какие метрики наиболее информативны при несбалансированных или многоклассовых данных?

**Для несбалансированных данных:**
- F1-score (macro, weighted)
- Precision-Recall AUC
- Matthews Correlation Coefficient (MCC)
- Cohen's Kappa
- Balanced Accuracy

**Для многоклассовых задач:**
- F1-score с различными стратегиями усреднения
- Log Loss
- Confusion Matrix (матрица ошибок)
- Classification Report

## 2. В чём разница между macro и weighted усреднением метрик в многоклассовой задаче?

**Macro-усреднение:**
- Вычисляет метрику для каждого класса отдельно и усредняет
- Все классы имеют одинаковый вес
- Чувствительно к качеству на малых классах
- Дает представление о среднем качестве по всем классам

**Weighted-усреднение:**
- Учитывает количество примеров в каждом классе
- Взвешивает вклад класса по его поддержке (support)
- Более репрезентативно для несбалансированных данных
- Отражает общее влияние на всю выборку

## 3. Почему accuracy может быть неинформативной при несбалансированных классах?

**Проблемы accuracy:**
- Может быть высокой при хорошем предсказании только мажоритарного класса
- Не отражает качество на миноритарных классах
- Модель может достигать 95% accuracy, игнорируя редкие классы
- Маскирует реальные проблемы модели

**Пример:**
- 95% примеров - класс A, 5% - класс B
- Модель, всегда предсказывающая A, имеет 95% accuracy
- Но совершенно бесполезна для класса B

## 4. Что измеряет log-loss и почему он полезен для оценки вероятностных моделей?

**Log-loss измеряет:**
- Расхождение между предсказанными вероятностями и истинными метками
- Штрафует за уверенные, но неправильные предсказания
- Учитывает не только правильность, но и уверенность модели

**Преимущества:**
- Чувствителен к качеству вероятностных предсказаний
- Позволяет оценить калиброванность вероятностей
- Более строгая и информативная метрика
- Учитывает неопределенность модели

## 5. Можно ли использовать ROC-AUC для задачи с тремя классами? Как это делается в scikit-learn?

- `multi_class='ovr'` (one-vs-rest): каждый класс против всех остальных

## 6. Какая метрика лучше отражает качество модели: F1-score или accuracy — и в каких случаях?

### F1-score предпочтительнее когда:
- **Несбалансированные классы** - accuracy может вводить в заблуждение
- **Важны обе ошибки** - и ложноположительные, и ложноотрицательные
- **Критические задачи** - медицина, безопасность, финансы
- **Миноритарные классы важны** - необходимо учитывать качество на редких классах
- **Нужен баланс** между precision и recall

### Accuracy подходит когда:
- **Сбалансированные классы** - равномерное распределение
- **Ошибки равнозначны** - стоимость FP и FN сравнима
- **Простая интерпретация** - понятность для нетехнической аудитории
- **Быстрая оценка** - первоначальный анализ модели

### Практическое правило:
- При дисбалансе классов > 80/20 выбирать F1-score
- Для сбалансированных данных можно использовать accuracy
- Всегда анализировать обе метрики в контексте задачи

## 7. Почему в многоклассовой классификации не всегда уместно использовать ROC-AUC?

### Ограничения ROC-AUC для многоклассовых задач:

**Интерпретационные сложности:**
- Усредненные значения маскируют проблемы отдельных классов
- Трудно определить, с какими классами возникают проблемы
- Один хороший класс может скрыть несколько плохих

**Вычислительные проблемы:**
- Экспоненциальный рост вычислений с увеличением классов
- Для K классов нужно построить K кривых (one-vs-rest) или K(K-1)/2 (one-vs-one)

**Смысловые ограничения:**
- ROC-AUC предполагает бинарную классификацию
- В многоклассовом случае метрика становится менее интуитивной
- Может не отражать реальную полезность модели

**Альтернативы:**
- Macro/Micro F1-score
- Precision-Recall по классам
- Confusion matrix анализ
- Per-class метрики

## 8. Как по матрице ошибок (confusion matrix) определить, с какими классами модель справляется хуже всего

### Шаги анализа confusion matrix:

**1. Анализ главной диагонали:**
- Низкие значения → проблемы с распознаванием класса
- Сравнение с ожидаемыми значениями (support классов)

**2. Построчные показатели (Recall):**
- Высокие значения вне диагонали в строке → модель пропускает класс
- Низкий recall для конкретного класса

**3. Постолбцовые показатели (Precision):**
- Высокие значения вне диагонали в столбце → много ложных срабатываний
- Низкий precision для класса

**4. Выявление паттернов путаницы:**
- Симметричные высокие значения между двумя классами → систематическая путаница
- Асимметричные паттерны → односторонние ошибки

**5. Количественные индикаторы проблем:**
- Низкий True Positive Rate для класса
- Высокий False Positive Rate для класса
- Низкая точность (precision) на классе
- Низкая полнота (recall) на классе

### Пример диагностики:
- Класс A: высокие значения в столбце A, но низкие в строке A → хорошая precision, плохой recall
- Класс B: высокие значения в строке B, но низкие в столбце B → хороший recall, плохая precision
- Классы C и D: взаимно высокие значения C→D и D→C → систематическая путаница
