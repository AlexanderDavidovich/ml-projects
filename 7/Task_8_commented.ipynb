# Сохранение и развертывание моделей scikit-learn

## 1. Чем joblib отличается от pickle при сохранении моделей scikit-learn?

**Joblib преимущества:**
- Оптимизирован для массивов NumPy (меньший размер файла)
- Более эффективное сжатие для объектов scikit-learn
- Рекомендован разработчиками scikit-learn
- Лучшая производительность для больших массивов

**Pickle:**
- Стандартная библиотека Python
- Универсальное решение для любых объектов
- Может создавать большие файлы для numpy-массивов
- Медленнее для объектов с большими числовыми данными

## 2. Что включает в себя сохранённый конвейер (pipeline): только веса модели или и предобработчик тоже?

**Сохраненный pipeline включает:**
- Все шаги предобработки (StandardScaler, Imputer, etc.)
- Обученные параметры каждого трансформера
- Саму модель классификации с весами
- Порядок и конфигурацию всех шагов
- Метаданные и параметры каждого компонента

## 3. Можно ли загрузить конвейер (pipeline), обученный в одной версии scikit-learn, в другой версии? Какие риски это несёт?

**Технически возможно, но с рисками:**
- **Несовместимость API** - изменения в методах и атрибутах
- **Разное поведение алгоритмов** - исправления багов могут изменить результаты
- **Изменения в трансформерах** - разные параметры по умолчанию
- **Ошибки десериализации** - невозможность загрузить объект

**Рекомендации:**
- Фиксировать версии зависимостей
- Тестировать на совместимость при обновлении
- Использовать виртуальные окружения

## 4. Почему важно сохранять конвейер (pipeline), а не только модель классификации?

**Критические причины:**
- **Согласованность предобработки** - идентичные преобразования при предсказании

- **Воспроизводимость** - гарантия одинаковых результатов
- **Автоматизация** - не требуется ручное применение трансформеров

## Контрольные вопросы

## 1. В чём принципиальное отличие логистической регрессии от линейной регрессии?

**Линейная регрессия:**
- Решает задачу регрессии (предсказание непрерывной величины)
- Использует линейную функцию: y = wX + b
- Минимизирует MSE (среднеквадратичную ошибку)
- Выходные значения не ограничены

**Логистическая регрессия:**
- Решает задачу классификации (бинарной или многоклассовой)
- Использует сигмоидную функцию для преобразования в вероятности
- Минимизирует log loss (логистическую потерю)
- Выходные значения в диапазоне [0,1] (вероятности)

## 2. Почему логистическая регрессия считается линейным классификатором, несмотря на нелинейную сигмоидную функцию?

**Линейность в пространстве признаков:**
- Решающая граница определяется линейной функцией: wX + b = 0
- Сигмоида только преобразует линейную комбинацию в вероятность
- Нелинейность сигмоиды не меняет линейность границы решения
- Классификация основана на знаке линейной комбинации

## 3. Как регуляризация помогает бороться с переобучением в логистической регрессии?

**L1-регуляризация (Lasso):**
- Штрафует сумму абсолютных значений коэффициентов
- Обнуляет неважные признаки (создает разреженность)
- Упрощает модель, отбирая наиболее значимые признаки

**L2-регуляризация (Ridge):**
- Штрафует сумму квадратов коэффициентов
- Уменьшает все коэффициенты, но не обнуляет
- Стабилизирует модель, снижая чувствительность к шуму

## 4. Почему важно использовать конвейер (pipeline) при машинном обучении? Какие типичные ошибки он предотвращает?

**Предотвращаемые ошибки:**
- **Утечка данных** - информация из тестовой выборки влияет на предобработку
- **Несогласованность** - разные преобразования для train/test
- **Невоспроизводимость** - разный порядок операций
- **Сложность deployment** - отдельное сохранение трансформеров и моделей

## 5. Какие метрики выбрать для оценки модели, если один из классов крайне редкий?

**Рекомендуемые метрики:**
- Precision-Recall AUC (вместо ROC-AUC)
- F2-score (если recall важнее precision)
- Matthews Correlation Coefficient (MCC)
- Average Precision
- Confusion matrix с акцентом на миноритарный класс

## 6. Какие из рассмотренных моделей масштабируются лучше всего на больших объёмах данных?

**Хорошо масштабируются:**
- Логистическая регрессия (стохастический градиентный спуск)
- Деревья решений (эффективные алгоритмы построения)
- Random Forest (параллельное обучение деревьев)

**Плохо масштабируются:**
- SVM (время обучения O(n²)-O(n³))
- kNN (требует хранения всех данных в памяти)

## 7. Какова роль кросс-валидации в процессе подбора гиперпараметров?

**Функции кросс-валидации:**
- Оценка обобщающей способности на разных подвыборках
- Снижение зависимости от конкретного разбиения данных
- Более надежная оценка качества моделей
- Выбор стабильных гиперпараметров

## 8. Почему нельзя подбирать гиперпараметры, глядя на тестовую выборку?

**Причины:**
- **Переобучение на тестовых данных** - модель подстраивается под тестовую выборку
- **Некорректная оценка обобщения** - теряется объективность оценки
- **Оптимистичные результаты** - завышенные метрики на известных данных
- **Нарушение принципа** - тестовая выборка должна быть полностью независимой

## 9. Какие шаги в конвейер (pipeline) можно добавить, если в данных есть пропущенные значения?

**Обработка пропущенных значений:**
- SimpleImputer (замена средним, медианой, модой)
- KNNImputer (замена на основе похожих наблюдений)
- IterativeImputer (многовариантная импутация)
- Добавление бинарных признаков пропусков

## 10. Как организовать автоматическое переобучение и обновление сохранённой модели при поступлении новых данных?

**Стратегии обновления моделей:**
- **Периодическое переобучение** - по расписанию на аккумулированных данных
- **Инкрементальное обучение** - онлайн-алгоритмы (SGD)
- **Ансамблирование** - добавление новых моделей в ансамбль
- **Мониторинг дрейфа данных** - автоматический запуск переобучения при ухудшении качества
