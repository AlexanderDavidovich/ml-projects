## Задача 1: Поиск 'Network Protocol' в RFC

```python
import urllib.request
url = 'https://www.rfc-editor.org/rfc/rfc-ref.txt'
filename = '../../data/rfc-ref.txt'
try:
    print('Загрузка...')
    urllib.request.urlretrieve(url, filename)
    print('Готово')
except Exception as e:
    print(f'Ошибка: {e}')
```

```python
!awk '/Network Protocol/ {print NR": "$0; count++} END {print "\nВсего:", count}' ../../data/rfc-ref.txt
```

## Задача 2: Параллелограмм

```python
%%writefile ../../data/parallelogram.awk
#!/usr/bin/awk -f
function area(a, h) { return a * h }
function perimeter(a, b) { return 2 * (a + b) }
function main(a, h, b) {
    printf "Площадь: %.2f\n", area(a, h)
    printf "Периметр: %.2f\n", perimeter(a, b)
}
BEGIN { main(10, 5, 7); main(15, 8, 10) }
```

```python
!awk -f ../../data/parallelogram.awk
```
